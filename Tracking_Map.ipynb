```python
from folium.plugins import AntPath
```

```python
track_map = folium.Map([Lat[0],Long[0]])

AntPath(trip_rt.values, delay = 800, dash_array = [10,20], color = "brown", 
        pulse_color = "yellow", weight = 5, opacity=1.2, tooltip = "<b>Multan to Fatehpur</b>",
        popup = folium.Popup("""
        <img src= "https://i0.wp.com/academiamag.com/wp-content/uploads/2022/12/h2-multan-primary-culture.jpg?fit=1310%2C760&ssl=1" 
        width = "450" height ="300" />
        <center><a> <font size="+1">Multan</font></a></center>
        """)).add_to(track_map)

track_map.save("Trip105.html")
```

```python
over_speed_loc=trip.loc[trip["over_speeding"]==1,["Latitude", "Longitude"]]
```

```python
over_speed_loc
```

```python
points=over_speed_loc.index
```

```python
points
```

```python
distance= pd.Series(points).diff().dropna().astype(int).values.tolist()
```

```python
distance
```

```python
final_lst=[]
lst=[]

for i,j in list(enumerate(distance)):
    if j == 1:
        lst.append(i)
    else:
        final_lst.append(lst)
        lst=[]
        
```

```python
final_lst
```

```python
os_lst_ind=[]

for i in final_lst:
    if len(i)>5:
        os_lst_ind.append(i)
```

```python
os_lst_ind
```
